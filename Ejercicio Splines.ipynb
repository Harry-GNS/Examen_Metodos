 Bajo qué condiciones se puede encontrar los splines cúbicos que cumplan con la condición de m=−3

Entonces la ecuacion que se debe modificar a mi criterio debe ser $S′_0(x_1)=S′_1(x_1)$



```python
import matplotlib.pyplot as plt
import numpy as np

def Spline(x: float, x0: float, pars: dict) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3

xs = [-1, 0, 1]
ys = [1, 5, 3]

# Coeficientes calculados para que S'_0(0) = S'_1(0) = -3
s = [
    {"a": 1.0, "b": 8.0, "c": 0.0, "d": -4.0},    # Primer tramo: de x=-1 a x=0
    {"a": 5.0, "b": -3.0, "c": -9.0, "d": 10.0},   # Segundo tramo: de x=0 a x=1
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 100)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="orange")

plt.scatter(xs, ys, color='blue', zorder=5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos (pendiente m=-3 en x=0)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
